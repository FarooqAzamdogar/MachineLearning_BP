# Scikit-learn basic Usage 


```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
#load dataset from seaborn
df= sns.load_dataset('tips')
df.head()
```

```python
sns.lmgplot(x='total_bill',y='tip',data=df)
```

```python
# selecting the features columns from df
X =  df[['total_bill']]
# selecting the label column from df
Y = df[['tip']]
```

```python
# load the linear regression model
model= LinearRegression()
# fit the model
model.fit(X, Y)
```

```python
# lets tip the  unknown and total_bill values
model.predict([[50]])
```
